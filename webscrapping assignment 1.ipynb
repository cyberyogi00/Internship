```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


page=requests.get("https://www.imdb.com/chart/top")
soup=BeautifulSoup(page.content)
movies=soup.find_all("td",class_="titleColumn")
ratings=soup.find_all("td",class_="ratingColumn imdbRating")

movie_list=[]
for i in range(50):
    movie_dict={}
    name=movies[i].find("a").get_text()
    rating=ratings[i].find("strong").get_text()
    year=movies[i].find("span").get_text()[1:-1]
    movie_dict["Name"]=name
    movie_dict["Year"]=year
    movie_dict["Rating"]=rating
    movie_list.append(movie_dict)

df=pd.DataFrame(movie_list)
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Scrape Top 10 ODI teams in men’s cricket
url = 'https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

team_names = []
matches = []
points = []
rating = []

for tr in soup.find_all('tr')[1:11]:
    td = tr.find_all('td')
    team_names.append(td[1].text.strip())
    matches.append(td[2].text.strip())
    points.append(td[3].text.strip())
    rating.append(td[4].text.strip())

odi_teams = pd.DataFrame({'Team Name': team_names,
                          'Matches': matches,
                          'Points': points,
                          'Rating': rating})

print("Top 10 ODI teams in men's cricket")
print(odi_teams)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


url = 'https://www.cnbc.com/world/?region=world'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

articles = soup.find_all('div', {'class': 'Card-titleContainer'})

headlines = []
times = []
links = []

for article in articles:
  
    headline = article.find('a').text.strip()
    headlines.append(headline)
    
for article in articles:
    time = article.find('time')
    times.append(time)
    
   
    link = article.find('a')['href']
    links.append(link)


news_df = pd.DataFrame({'Headline': headlines,'Time': times,'Link': links})

print(news_df)
```
